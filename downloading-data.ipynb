# DATA MINING

### funcao para minerar os dados

```python
#criando as querys

def cria_querys(restaurantes, emojis):
    querys = []
    
    for restaurante in restaurantes:
        for emoji in emojis:
            querys.append(f"{restaurante} {emoji}")
    
    return querys
```

```python
querys_positivos = cria_querys(restaurantes, sentimentos_positivos)
querys_negativos = cria_querys(restaurantes, sentimentos_negativos)
```

```python
def data_search(limit: int, list_of_querys: list) -> list:
    tweets = []
    
    # buscando as querys
    for query in list_of_querys:
        
        for tweet in sntwitter.TwitterSearchScraper(query +" lang:pt").get_items():
            
            if len(tweets) >= limit:
                print("A quantidade de tweets requisitados foi alcanÃ§ada")
                break;
            
            else:
                tweets.append({"date":tweet.date, "user_name" :tweet.user.username,
                               "tweet":tweet.content})
    return tweets
```

```python
tweets_positivos = data_search(3000, querys_positivos)
tweets_negativos = data_search(3000, querys_negativos)
```

## funcao para salvar os dados

```python
def cria_data_frame(dados: list) -> pd.DataFrame:
    texto, nome_usuario, data_publicacao = [], [], []

    for tweet in dados:
        texto.append(tweet["tweet"])
        data_publicacao.append(tweet["date"])
        nome_usuario.append(tweet["user_name"])
    
    main_data = pd.DataFrame({"user_name":nome_usuario, "tweet":texto,
                           "date":data_publicacao})
    
    return main_data
```

```python
df_positivo = cria_data_frame(tweets_positivos)
df_negativo = cria_data_frame(tweets_negativos)
```

### salvando dados

```python
df_positivo.to_csv("raw-data/coleta-positivo-1.csv", sep="\t", index=False)
df_negativo.to_csv("raw-data/coleta-negativo-1.csv", sep="\t", index=False)
```
